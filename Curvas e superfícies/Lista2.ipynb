## Questão 2)
Curva escolhida: Espiral de Arquimedes \
Equação parametrizada: $(t*cos(t), t*sin(t))$, para $t\in\mathbb{R}^+$ \
A Espiral de Arquimedes recebeu esse nome por conta do matemático Arquimedes de Siracusa. Ela é calculada a partir de uma sucessão de pontos equidistantes que se afastam a uma taxa constante do ponto de origem enquanto a reta gira a uma velocidade também constante. A curva de Arquimedes tem a propriedade de manter constante a distância entre duas curvas sucessivas. \
Para $t>0$ temos uma curva percorrida no sentido anti-horário a partir da origem, enquanto para $t<0$ temos uma curva simétrica à $t>0$, percorrida no sentido horário.\
Vamos plotar essa curva.


```python
figures = []

curve(t) = (t*cos(t), t*sin(t))
gamma = parametric_plot(curve, (t, 0, 10), color = 'green')
d_alpha = diff ( curve , t )
listaa = []
contador = 0
for i in range(0,10):
    listaa.append(contador)
    contador = contador+1
for i in listaa:
    vector = plot ( d_alpha . subs ( t = i) , start = curve . subs ( t = i) , color ='#962B09', ymin = -15, ymax = 15)
    figures.append(gamma + vector)
a = animate(figures, xmax=15, xmin = -15)
a.show()

```

Note que a medida que a espiral se afasta da origem o vetor velocidade é maior. Podemos realizar uma reparametrização por comprimento de arco pra uniformizar a forma pela qual o vetor tangente percorre a curva.\
A reparametrização por comprimento de arco é uma reparametrização que permite, na junção dos vetores tangentes, originarem uma nova curva exatamente igual a original e portanto útil no cálculo do comprimento de arco e também no percorrer da curva.
Conforme os passos abaixo:\
1- Calcular a derivada da curva:
$$curve(t) = (t*cos(t), t*sin(t))$$
$$curve'(t) = (cos(t) - t*sin(t),sin(t) + t*cos(t))$$
2- Calcular o módulo da derivada:
$$||curve'(t)||^2 = cos^2(t) \cancel{- 2*t*sin(t)cos(t)} + t^2sin^2(t) + sin^2(t) \cancel{+ 2*t*sin(t)cos(t)} + t^2*cos^2(t)$$
$$||curve'(t)||^2 = cos^2(t)+ sin^2(t)  + t^2(sin^2(t) + cos^2(t))$$
$$||curve'(t)||^2 = 1 + t^2$$
3- Resolver a integral do módulo de 0 até $t$:
$$\int_0^t 1 + t^2 dt = t + t^3/3$$
4- Escrever $t$ em função de $s$ (foram usados métodos computacionais aqui):
$$s = t + t^3/3$$
$$3*s = 3*t + t^3$$
$$t = ((9*s^2 + 4)^{(1/2)} + 3*s)^{(1/3)}/2^{(1/3)} - 2^{(1/3)}/((9*s^2 + 4)^{(1/2)} + 3*s)^{(1/3)}$$
5- Substituir na equação original (vamos omitir aqui por por ter origindo uma expressão muito grande).

Vamos plotar essa curva reparametrizada por comprimento de arco.\
Note o percorrer da curva com velocidade uniforme.

```python
s = var('s')
figures = []
curve(s) = ((((9*s^2 + 4)^(1/2) + 3*s)^(1/3)/2^(1/3) - 2^(1/3)/((9*s^2 + 4)^(1/2) + 3*s)^(1/3))*cos(((9*s^2 + 4)^(1/2) + 3*s)^(1/3)/2^(1/3) - 2^(1/3)/((9*s^2 + 4)^(1/2) + 3*s)^(1/3)), (((9*s^2 + 4)^(1/2) + 3*s)^(1/3)/2^(1/3) - 2^(1/3)/((9*s^2 + 4)^(1/2) + 3*s)^(1/3))*sin(((9*s^2 + 4)^(1/2) + 3*s)^(1/3)/2^(1/3) - 2^(1/3)/((9*s^2 + 4)^(1/2) + 3*s)^(1/3)))
gamma = parametric_plot(curve, (s, 0, 300), color = 'green')
d_alpha = diff ( curve , s )
listaa = []
contador = 0
for i in range(0,300):
    listaa.append(contador)
    contador = contador+1
for i in listaa:
    vector = plot ( d_alpha . subs ( s = i) , start = curve . subs ( s = i) , color ='#962B09', ymin = -10, ymax = 10)
    figures.append(gamma + vector)
a = animate(figures, xmax=10, xmin = -10)
a.show()
```
