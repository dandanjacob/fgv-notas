# Questão 1) 
Precisamos de duas funções que variem de $0$ à $1$: $cos(x)$ e $sin(x)$.<br>
 Se o intervalo for $(pi/2 ; 5pi/2)$, a curva começa em $(0,1)$ e acaba no mesmo ponto.<br>
Em seguida, inserirmos vetores velocidade, através da função $diff$. <br>
Por fim, geramos uma imagem para cada $t$ previamente selecionado e incluimos essa imagem na lista $figures$, que será animada.


```python
figures = [] #lista de figuras vazia
curve_alpha(s) = (cos(s), sin(s)) #curva parametrizada
d_alpha = diff ( curve_alpha , s ) #diferencial da curva parametrizada
circle = parametric_plot ( curve_alpha , (s ,pi/2 , 5*pi/2 ),  thickness =2, ymin=-2, ymax=2) #plotagem do círculo 
setas = [pi/2, pi, 3*pi/2, 2*pi] #lista de 't's
for i in setas: #para cada 't', um vetor tangente 
    vector = plot ( d_alpha . subs ( s = i) , start = curve_alpha . subs ( s = i) , color ='#962B09',ymin=-2, ymax=2) #vtor tangente do 't'_i
    figures.append(circle + vector) #inclusão na lista de imagens do círculo e do vetor i
a = animate(figures,xmax = 2, xmin = -2) #animação da lista de figuras
a.show() #mostra gif
```

# Questão 7) 
=>)Se $||\alpha'||$ é constante, então $||\alpha'(t)^2|| = <\alpha'(t),\alpha'(t)>$ também é constante. Como a derivada de uma constante é zero, $\frac{d}{dx}< \alpha'(t),\alpha'(t)> = 0$. <br>
Sendo  $$\frac{d}{dx}<\gamma(t),\lambda(t)> = <\gamma'(t),\lambda(t)> + <\gamma(t)\lambda'(t)>,$$ temos:

$$ \frac{d}{dx}<\alpha'(t),\alpha'(t)> = <\alpha''(t),\alpha'(t)>+<\alpha'(t),\alpha''(t)> = 0 $$
Daí, $$2<\alpha''(t), \alpha'(t)> \implies <\alpha''(t), \alpha'(t)> = 0$$.Como a curva é regular, $\alpha'(t)\neq0$.
Logo, se o produto interno entre $\alpha'(t)$ e $\alpha''(t)$ é zero, conclui-se que eles são ortogonais.
<=) Avaliando o ângulo entre $||\alpha''(t)||$ e $||\alpha'(t):$ \
$$cos\theta = \frac{<\alpha''(t), \alpha'(t)>}{||\alpha'(t)||||\alpha''(t)||}$$
$$cos\theta||\alpha'(t)||||\alpha''(t)|| = 0$$
Nos resta apenas $cos\theta=0$, portanto os vetores $\alpha'(t)$ e $\alpha''(t)$ são ortogonais.

# Questão 8
Como visto em aula, para uma curva ser diferenciável é necessário que o vetor velocidade seja diferente de zero para todo $t \in \mathbb{R}$, segue que 
$$\frac{d}{dx}(x(t),y(t)) = (x'(t), y'(t))$$
Como pelo enunciado, $x'(t) \neq 0$, então $\alpha'(t) \neq 0$.
Para sermos mais precisos, vamos demonstrar que $\forall x \in x(I)=\{x|(x(t),y(t)) \in \alpha(t)$ e $x(t)=x\} \exists$ um único $y$ associado a ele em $\alpha$. 
Sejam $(x(t_1),y(t_1))$ e $(x(t_2),y(t_2))$ dois pontos distintos da curva.
 Pelo Teorema do Valor Médio temos 
$\frac{x(t_1)-x(t_2)}{(t_1-t_2)} = x'(t_3) \neq 0 t_3 \in (t_1,t_2)$
 Portanto, para qualquer $t_1$ e $t_2$ temos $x(t_1)\neq x(t_2)$. Como a curva não admite dois pontos com mesma coordenada nas abcissas, podemos associar $f:x(I) \longmapsto y(I)$ definida de modo que $f(x) =y$.
Seja agora $x\in x(I)$ e $h$ de modo que $x+h \in x(I)$. Sendo $x+h=x(t_h)$ e $x=x(t)$:
$ \lim_{h\longmapsto 0}\frac{f(x+h)-f(x)}{x+h-x} = \lim_{h\longmapsto 0}\frac{y(t_h)-y(t)}{x(t_h)-x_(t)} = \lim_{h\longmapsto 0}\frac{\frac{y(t_h)-y(t)}{t_h-t}}{\frac{x(t_h)-x_(t)}{t_h-t}}$.
Provamos então que $x(t)$ é injetiva em $I$ e portanto $x:I \longmapsto x(I)$ é bijetiva e inversível. Por fim, como $h \longmapsto 0$, então $t_h \longmapsto t$ e $f'(x) = \frac{y'(t)}{x'(t)}$, o que prova $f$ ser derivável.

$γ(t) = (e^t*cos(t), e^t*sin(t))$
Sendo <b>u</b> o vetor da curva e <b>v</b>,  o ângulo entre os vetores corresponde a 
$$cos(\theta) = \frac{<u,v>}{||u||.||v||}.$$
O produto interno entre <b>u</b> e <b>v</b> é $e^{2t}$, $||u|| = e^t$ e  $||v|| = \sqrt{2}e^t$.
Sendo assim, 
$$ cos(\theta) = \frac{(e^{2t})}{e^t\sqrt{2}e^t} $$
Logo, $cos(\theta) = \frac{1}{\sqrt{2}}$, independente de $t$.
